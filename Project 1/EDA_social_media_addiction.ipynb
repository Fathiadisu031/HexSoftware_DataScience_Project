```python
data1   = data.drop(['Conflicts_Over_Social_Media', 'Country'], axis=1) # to drop the columns from the dataset]
data1.head()
```

```python
data1.info() # to check the data types of the columns
```

```python
# univariate analysis
```

```python
# Icluding only numerical columns for correlation analysis
numerical_cols = data1.select_dtypes(
    include=['int64', 'float64']
    ).columns.tolist()
numerical_cols.remove('Student_ID') # Exclude ID from general numerical analysis

```

```python
# Histogram for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(12, 6))
    sns.histplot(data1[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
    
```

```python
# Boxplot for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=data1, y=col)
    plt.title('Box plot for ' + col)
    plt.ylabel(col)
```

```python
# Categorical analysis
categorical_cols = data1.select_dtypes(include=['object']).columns.tolist()
print(categorical_cols)
```

```python
for col in categorical_cols:
        data1[col].value_counts().plot(kind = 'bar', figsize = (12,6))
        plt.title(f'Count of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()
```

```python
# relationship analysis
```

```python
# To group the age into ranges
data1['Age_Group'] = pd.cut(data1
                            ['Age'],
                            bins=[18, 20, 22, 24, 25], # Define your age ranges
                            labels=['18-19', '20-21', '22-24', '25+']) # Define your age group labels
data1['Age_Group'].value_counts() # to check the value counts of the Age_Group column
```

```python
#scatterplot to visualize the relationship between Addicted_Score and Mental_Health_Score
plot = sns.relplot(x = 'Mental_Health_Score', y = 'Addicted_Score', data = data1, kind = 'scatter')

```

```python
# to check the correlation of the dataset
correlation = data1[numerical_cols].corr(numeric_only=True)
print(correlation)
```

```python
#Heatmap to visualize the correlation
sns.heatmap(correlation,
            xticklabels = correlation.columns,
            yticklabels = correlation.columns,
            annot = True)
```

```python
#pairplot to visualize the relationship between the columns
sns.pairplot(data1)
```

```python
# Multiple Box Plot: Avg_Daily_Usage_Hours by Most_Used_Platform
plt.figure(figsize=(12, 7))
sns.boxplot(x='Most_Used_Platform', y='Avg_Daily_Usage_Hours', data = data1)
plt.title('Average Daily Usage Hours by Most Used Platform')
plt.xlabel('Most Used Platform')
plt.ylabel('Average Daily Usage Hours')
plt.show()
```

```python
# Multiple Box Plot: Mental_Health_Score by Academic_Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='Academic_Level', y='Mental_Health_Score', data = data1)
plt.title('Mental Health Score by Academic Level')
plt.xlabel('Academic Level')
plt.ylabel('Mental Health Score')
plt.show()
```

```python

# Stacked bar chart for Gender and Affects_Academic_Performance
gender_academic_crosstab = pd.crosstab(data1['Age_Group'], data1['Affects_Academic_Performance'], normalize='index') # normalize to show proportions
gender_academic_crosstab.plot(kind='bar', stacked=True, figsize=(8, 5))
plt.title('Impact on Academic Performance by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion')
plt.xticks(rotation=0)
plt.legend(title='Affects Academic Performance')
plt.show()
```
